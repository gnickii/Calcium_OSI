```python
import pandas as pd
import numpy as np
from collections import Counter
```

```python
# Cargar archivo registro arduino y encontrar tiempo en ms y banderas
fn_reg = r'C:\Users\neuro\OneDrive\Documentos\Nicole\BTBR_GCaMP6\BTBR_R2_HEMBRA_PASIVA_27-09-2023\BTBR_R2_HEMBRA_PASIVA_reg122535.txt'
with open(fn_reg, 'r') as file:
    C = file.readlines()
C
```

```python
# Procesar las líneas del archivo
datasplit = [line.split('.') for line in C[1:]]  # Split lines starting from the second line
timestamp_flgs = [str(item[0]) for item in datasplit]
millisec = np.array([float(item[2]) for item in datasplit])
sensores = np.array([float(item[3]) for item in datasplit])
print(millisec)
print(sensores)
```

```python
# Encontrar banderas únicas y contar ocurrencias
flags, counts = np.unique(sensores, return_counts=True)
value_counts = np.column_stack((flags, counts))
print(value_counts)
```

```python
# Cargar archivo registro psychopy y encontrar estímulos
fn_log = r'C:\Users\neuro\OneDrive\Documentos\Nicole\BTBR_GCaMP6\BTBR_R2_HEMBRA_PASIVA_27-09-2023\BTBR_R2_HEMBRA_PASIVA_log122548.txt'
with open(fn_log, 'r') as file:
    C_log = file.readlines()
C_log
```

```python
# Procesar las líneas del archivo de psychopy
datasplit_log = [line.split('.') for line in C_log]
tstamp_stlog = [str(item[0]) for item in datasplit_log]
estimulos = np.array([float(item[2]) for item in datasplit_log])
```

```python
# Obtener identidades únicas de cada estímulo
stims, stim_counts = np.unique(estimulos, return_counts=True)
value_counts_st = np.column_stack((stims, stim_counts))
print(stims)
```

```python
# Filtrar estímulos
stims = stims[stims != 11]
# Convertir a entero el valor usado para definir la dimensión del array
idx_t_stim = np.zeros((int(value_counts_st[0, 1]), len(stims)))
print(stims)
```

```python
for s, stim in enumerate(stims):
    t_stamp_idx = np.where(estimulos == stim)[0]
    t_stamp = [tstamp_stlog[i - 1] for i in t_stamp_idx]

    for j, ts in enumerate(t_stamp):
        hms = list(map(int, ts.split(':')))
        hms[2] += 1
        t_stamp2 = f"{hms[0]:02}:{hms[1]:02}:{hms[2]:02}"

        idx_timebin = np.concatenate((
            np.where(np.array(timestamp_flgs) == ts)[0],
            np.where(np.array(timestamp_flgs) == t_stamp2)[0]
        ))

        if len(idx_timebin) > 0:
            idx_stim_j = np.where(sensores[idx_timebin[0]:idx_timebin[-1]] == 4)[0] + idx_timebin[0]
            if len(idx_stim_j) > 0:
                idx_t_stim[j, s] = idx_stim_j[0]

print(t_stamp_idx)  # Para verificar el resultado
```

```python
# Alinear tiempo 0 a la adquisición del primer frame
millisec_f0 = millisec - millisec[0]

framt = np.where(sensores == 23)[0]
t_fram = millisec_f0[framt]
ll = np.diff(t_fram)
idx_t_fr = np.where(ll > 100)[0] + 1
t_fram_clean = np.insert(t_fram[idx_t_fr], 0, t_fram[0])
framt
```

```python
# Encontrar frames en los que iniciaron los estímulos
t_Stimulus_ms = millisec_f0[idx_t_stim.astype(int)]
tf_stimulus = np.round(t_Stimulus_ms / 200).astype(int)
# Corrección del frame correcto
a = t_Stimulus_ms - t_fram_clean[tf_stimulus] > 200
idx_t_adj = np.where(a)[0]
tf_stimulus[idx_t_adj] += 1
print(tf_stimulus)
```

```python
tf_stimulus=tf_stimulus.T
# Reshape stims to match idx_t_stim's rows
stims_reshaped = stims[:, np.newaxis]

# Create a 3D array by stacking stims_reshaped with idx_t_stim along a new axis
stimsnframes = np.stack([stims_reshaped] * tf_stimulus.shape[1], axis=1)

# Concatenate stimsnframes with idx_t_stim along the last axis
result = np.concatenate((stimsnframes, tf_stimulus[:, :, np.newaxis]), axis=2)

print(result.shape)  
print(result)
```
