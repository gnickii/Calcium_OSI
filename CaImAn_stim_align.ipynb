```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
f_df_traces=r'C:\Users\neuro\OneDrive\Documentos\Nicole\BTBR_GCaMP6\BTBR_R2_HEMBRA_PASIVA_27-09-2023\C_traces_BTBR_R2_HEMBRA_PASIVA.csv'
C=pd.read_csv(f_df_traces)
C
```

```python
# Find the length of the longest column (i.e., the number of rows in the DataFrame)
max_len = len(C)

# Add a new column with values from 1 to max_len
C.insert(0, 'Frame', range(0, max_len ))

# Display the DataFrame with the new column
C
```

```python
stimsnframes = np.load('result.npy')

print(stimsnframes)
stimsnframes.shape
```

```python
# Extraer las segundas columnas de stimsnframes
stim_frame = stimsnframes[:, :, 1].flatten()

# Encontrar las filas donde 'Time Frame' coincide con los valores de la segunda columna de stimsnframes
C_stim_frames = C[C['Frame'].isin(stim_frame)].copy()

# Extraer la primera columna de stimsnframes para los valores coincidentes
tf_stimulus = []

for time_frame in C_stim_frames['Frame']:
    # Encontrar el índice correspondiente en stimsnframes
    index = np.where(stimsnframes[:, :, 1] == time_frame)[0][0]  # Busca el índice de la fila en stimsnframes
    # Extraer el valor correspondiente de la primera columna
    tf_stimulus.append(stimsnframes[index, 0, 0])

# Usar .loc[] para añadir la nueva columna sin el warning
C_stim_frames.loc[:, 'Stim'] = tf_stimulus

# Reordenar las columnas para que 'Stim Value' esté al lado de 'Time Frame'
cols = ['Frame', 'Stim'] + [col for col in C_stim_frames.columns if col not in ['Frame', 'Stim Value']]
C_stim_frames = C_stim_frames[cols]

# Mostrar el resultado

C_stim_frames
```

```python
neuronas = ['1', '2', '3']  # Cambia esto a las columnas que deseas graficar
time = C['time']
Estimulo = C_stim_frames[['time', 'Stim']].values

# Asumimos que cada estímulo dura aproximadamente 4 segundos
duracion_estimulo = 4

# Agregar una columna de tiempo final para cada estímulo
C_stim_frames['End_time'] = C_stim_frames['time'] + duracion_estimulo

# Configurar las variables para graficar
neuronas = ['1', '2', '3']
time = C['time']

# Colores para los diferentes tipos de estímulos
colores_estimulos = {
    0.0: 'red',
    45.0: 'blue',
    90.0: 'green',
    135.0: 'purple',
    180.0: 'orange',
    225.0: 'pink',
    270.0: 'brown',
    315.0: 'gray'
}

num_columnas = len(neuronas)
fig, axs = plt.subplots(num_columnas, 1, figsize=(12, 3 * num_columnas), sharex=True)

# Crear un conjunto para las entradas de la leyenda
legend_handles = {}
for _, row in C_stim_frames.iterrows():
    start = row['time']
    end = row['End_time']
    stimulus = row['Stim']
    if isinstance(stimulus, pd.Series):
        stimulus = stimulus.values[0]
    color = colores_estimulos.get(stimulus, 'black')
    if stimulus not in legend_handles:
        legend_handles[stimulus] = plt.Line2D([0], [0], color=color, lw=4, label=f'Estímulo {int(stimulus)}°')

# Graficar cada columna en su propio subgráfico
for i, columna in enumerate(neuronas):
    axs[i].plot(time, C[columna], label=f'Neurona {columna}')
    # Añadir las barras de color para los intervalos de estímulo
    for _, row in C_stim_frames.iterrows():
        start = row['time']
        end = row['End_time']
        stimulus = row['Stim']
        if isinstance(stimulus, pd.Series):
            stimulus = stimulus.values[0]
        color = colores_estimulos.get(stimulus, 'black')
        axs[i].axvspan(start, end, color=color, alpha=0.3)
    
    axs[i].set_ylabel(f'Neurona {columna}')
    axs[i].grid(True)

# Añadir leyenda para estímulos al primer subgráfico
axs[0].legend(handles=legend_handles.values(), loc='upper right')

# Ajustar el diseño para que no se superpongan
plt.xlabel('Tiempo (s)')
plt.tight_layout()
plt.savefig('trazos_con_estimulos.png', bbox_inches='tight')
plt.show()
```
