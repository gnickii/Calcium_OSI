```python
# Find the length of the longest column (i.e., the number of rows in the DataFrame)
max_len = len(C)

# Add a new column with values from 1 to max_len
C.insert(0, 'Frame', range(0, max_len ))

##
times = [i * 0.200 for i in range(max_len)]


C.insert(0, 'Time', times)

# Display the DataFrame with the new column
C
```

```python
stimsnframes = np.load('result.npy')

print(stimsnframes)
stimsnframes.shape
```

```python
# Extraer las segundas columnas de stimsnframes
stim_frame = stimsnframes[:, :, 1].flatten()

# Encontrar las filas donde 'Frame' coincide con los valores de la segunda columna de stimsnframes
C_stim_frames = C[C['Frame'].isin(stim_frame)].copy()

# Extraer la primera columna de stimsnframes para los valores coincidentes
tf_stimulus = []

for time_frame in C_stim_frames['Frame']:
    # Encontrar el índice correspondiente en stimsnframes
    index = np.where(stimsnframes[:, :, 1] == time_frame)[0][0]  # Busca el índice de la fila en stimsnframes
    # Extraer el valor correspondiente de la primera columna
    tf_stimulus.append(stimsnframes[index, 0, 0])

# Elimina la columna 'Stim' si ya existe
if 'Stim' in C_stim_frames.columns:
    C_stim_frames = C_stim_frames.drop(columns=['Stim'])

# Añadir la nueva columna 'Stim'
C_stim_frames['Stim'] = tf_stimulus

# Reordenar las columnas para que 'Stim' esté al lado de 'Frame'
cols = ['Frame', 'Stim'] + [col for col in C_stim_frames.columns if col not in ['Frame', 'Stim']]
C_stim_frames = C_stim_frames[cols]

# Mostrar el resultado
C_stim_frames
```

```python
# Crear una lista de estímulos únicos
estimulos_unicos = C_stim_frames['Stim'].dropna().unique()

duracion_basal = 2

neuronas = [col for col in C.columns if isinstance(col, int) or col.isdigit()]
neuronas = [int(col) for col in neuronas]

# Diccionario para almacenar qué neuronas están moduladas por cada estímulo
neuronas_moduladas = {stimulus: [] for stimulus in estimulos_unicos}

# Para cada estímulo único, analizar cada neurona
for stimulus in estimulos_unicos:
    # Filtrar los intervalos de estímulo para el estímulo actual
    C_stim_frames_stimulus = C_stim_frames[C_stim_frames['Stim'] == stimulus]

    print(f"\nAnálisis para el estímulo {int(stimulus)}:")
    
    for i, columna in enumerate(neuronas):
        # Verificar actividad basal y durante el estímulo para cada intervalo del estímulo
        modulado = False
        for _, row in C_stim_frames_stimulus.iterrows():
            start_stim = row['Time']
            end_stim = row['End_time']
            
            # Filtrar la actividad basal (2 segundos antes del estímulo)
            basal_mask = (time >= (start_stim - duracion_basal)) & (time < start_stim)
            actividad_basal = C.loc[basal_mask, columna]
            
            # Calcular media y desviación estándar de la actividad basal
            media_basal = actividad_basal.mean()
            std_basal = actividad_basal.std()
            
            # Filtrar la actividad durante el estímulo
            stim_mask = (time >= start_stim) & (time < end_stim)
            actividad_estimulo = C.loc[stim_mask, columna]
            
            # Verificar si la actividad supera 3 veces la desviación estándar
            if (actividad_estimulo > (media_basal + 3 * std_basal)).any():
                modulado = True
                print(f"  Neurona {columna} modulada (supera 3 SD)")
        
        # Marcar si la neurona está modulada por este estímulo
        if modulado:
            neuronas_moduladas[stimulus].append(columna)

# Imprimir las neuronas moduladas por estímulo
for stimulus, neuronas in neuronas_moduladas.items():
    print(f"Para el estímulo {int(stimulus)}, las neuronas moduladas son: {', '.join(map(str, neuronas)) if neuronas else 'Ninguna'}")
```
